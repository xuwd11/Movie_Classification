```python
NUM_CLASSES=5
from torchvision.models import resnet152
from torch import nn
CNN_base = resnet152(pretrained=True)

# Freeze base layers
for child in list(CNN_base.children())[:-3]:
    for param in child.parameters():
        param.requires_grad = False

# Conv4, turn off grad in first 2 blocks
for child in list(list(CNN_base.children())[-3].children())[:2]:
    for param in child.parameters():
        param.requires_grad = False

CNN = nn.Sequential(*list(CNN_base.children())[:-1])
CNN = nn.Sequential(CNN, nn.Linear(in_features=2048,out_features=NUM_CLASSES))
```

```python
import numpy as np
model_parameters = filter(lambda p: p.requires_grad, CNN.parameters())
params = sum([np.prod(p.size()) for p in model_parameters])
params
```
